```python
import pandas as pd
import numpy as np
import re
```

```python
selected_cols = ['name', 'county_name', 'city_name', 'aland', 'median_household_income', 'average_household_income', 'per_capita_income', 'percent_of_households_with_incomes_200000_or_more',
                'householder_under_25_years', 'householder_25_to_44_years', 'householder_45_to_64_years', 'householder_65_years_and_over'
                , 'hhi_total', 'hhi_lt_25k', 'hhi_25k_to_49k', 'hhi_50k_to_749k', 'hhi_75k_to_999k',
                'hhi_100k_to_1490k', 'hhi_150k_to_1999k', 'hhi_200k_or_more', 'race_and_ethnicity_total', 
                'population_density_sq_mi', 'median_age', 'educational_attainment_bachelors', 'educational_attainment_graduate',
                'family_poverty_pct', 'unemployment_pct', 'housing_units', 'median_value_of_owner_occupied_units']
raw = pd.read_csv("/Users/junfengli/Documents/Stat170A/Project/California_IncomeByZipDemographics/California_IncomeByZipDemographics.csv", skiprows = 4, usecols = selected_cols)
```

```python
raw
```

```python
raw = raw[raw['householder_under_25_years'] != "-US$1"]
raw_filtered = raw[raw['median_household_income'] != "-US$1"]
```

```python
numeric_conversion_list = ['aland', 'median_household_income', 'average_household_income', 'per_capita_income',
                'householder_under_25_years', 'householder_25_to_44_years', 'householder_45_to_64_years', 'householder_65_years_and_over'
                , 'hhi_total', 'hhi_lt_25k', 'hhi_25k_to_49k', 'hhi_50k_to_749k', 'hhi_75k_to_999k',
                'hhi_100k_to_1490k', 'hhi_150k_to_1999k', 'hhi_200k_or_more', 'race_and_ethnicity_total', 
                  'housing_units', 'median_value_of_owner_occupied_units']
for x in numeric_conversion_list:
    raw_filtered[x] =  [re.sub(r'[^0-9]+','', str(y)) for y in raw_filtered[x]]
    
raw_filtered['population_density_sq_mi'] =  [re.sub(r',','', str(y)) for y in raw_filtered['population_density_sq_mi']]
```

```python
raw_filtered
```

```python
zipcode_list = []
for index, row in raw_filtered.iterrows():
    if int(row['race_and_ethnicity_total']) >= 8000:
        zipcode_list.append(row['name'])
with open("zipcode_list.txt", 'w+') as output:
    for x in zipcode_list:
        output.write(str(x) + '\n')
```

```python
raw_filtered.to_csv('demographic.csv', index = False)
```
