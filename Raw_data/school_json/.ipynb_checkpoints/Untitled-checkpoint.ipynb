```python
import csv

def create_csv(p, head: list):
    path = p
    with open(path, "w") as f:
        csv_write = csv.writer(f)
        csv_head = head
        csv_write.writerow(csv_head)


def write_csv(p, data: list):
    with open(p, "a+") as f:
        csv_write = csv.writer(f)
        csv_write.writerow(data)
```
