```python
import csv

def create_csv(p, head: list):
    path = p
    with open(path, "w") as f:
        csv_write = csv.writer(f)
        csv_head = head
        csv_write.writerow(csv_head)


def write_csv(p, data: list):
    with open(p, "a+") as f:
        csv_write = csv.writer(f)
        csv_write.writerow(data)
```

```python
head = ['schoolid','schoolName','url','street_address','city','zip',
       'lowGrade','highGrade','schoolLevel',"isCharterSchool","isMagnetSchool",
       "isVirtualSchool",'isTitleISchool','districtName','countyName','rank',
        "rankOf",'rankStars','rankLevel','rankStatewidePercentage','averageStandardScore','rankMovement',
        'numberOfStudents','teachersFulltime','isPrivate'
       ]
```

```python
create_csv('school_general.csv',head)
```

```python
def concatecsv(p,file):
    test = open(file)
    content = test.readlines()
    j = json.loads(content[0])
    for school in j['schoolList']:
        print(school['schoolid'])
        result = []
        result.append(school['schoolid'])
        result.append(school['schoolName'])
        result.append(school['url'])
        result.append(school['address']['street'])
        result.append(school['address']['city'])
        result.append(school['address']['zip'])
        result.append(school['lowGrade'])
        result.append(school['highGrade'])
        result.append(school['schoolLevel'])
        result.append(school['isCharterSchool'])
        result.append(school['isMagnetSchool'])
        result.append(school['isVirtualSchool'])
        result.append(school['isTitleISchool'])
        if 'district' in school.keys():
            result.append(school['district']['districtName'])
        else:
            result.append('')
        result.append(school['county']['countyName'])
        try:
            result.append(school['rankHistory'][0]['rank'])
            try:
                result.append(school['rankHistory'][0]['rankOf'])
            except KeyError:
                result +=['']
            result.append(school['rankHistory'][0]['rankStars'])
            result.append(school['rankHistory'][0]['rankLevel'])
            try:
                result.append(school['rankHistory'][0]['rankStatewidePercentage'])
            except KeyError:
                result +=['']
            result.append(school['rankHistory'][0]['averageStandardScore'])
        except TypeError:
            result +=['']*6
        result.append(school['rankMovement'])
        if 'district' in school.keys() and school['schoolYearlyDetails'] != []:
            result.append(school['schoolYearlyDetails'][0]['numberOfStudents'])
            result.append(school['schoolYearlyDetails'][0]['teachersFulltime'])
        else:
            result += ['']*2
        result.append(school['isPrivate'])
        print(len(result))
        write_csv(p,result)
    test.close()
```

```python
def interate_file(files):
    for p in files:
        if '.json' in p:
            print(p)
            concatecsv('school_general.csv',p)
```
