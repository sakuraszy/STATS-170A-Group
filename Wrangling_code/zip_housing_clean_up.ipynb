```python
import matplotlib.pyplot as plt
from matplotlib import colors
import pandas as pd
import seaborn as sns
import numpy as np

from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn import cluster, datasets
```

```python
housing = pd.read_csv("../Data_raw/zip_housing.csv", index_col = 16)
```

```python
housing
```

```python
housing[['baths_half']] = housing[['baths_half']].fillna(0)
housing
```

```python
housing[['baths_full']] = housing[["baths_full"]].fillna(housing.mean()).astype({"baths_full": int})
housing['baths'] = housing['baths_full'] + housing['baths_half']*0.5
housing
```

```python
##housing = housing.fillna(housing.mean()).astype({"year_built": int, "beds": int, "baths_full": int, "baths_half": int, "garage": int,"price": int,
##                         "lot_size": int,"building_size": int, "postal_code": str})
```

```python
housing_grouped = housing.groupby('postal_code')
count = 0
df = pd.DataFrame()
for name, group in housing_grouped:
    ##print(len(group))
    group.fillna(group.median(), inplace = True)
    ##print(group)
    ##truncate_num = int(len(group)*0.05)
    ##truncate_len = len(group) - truncate_num
    ##group.sort_values(by=['price'])
    ##group.truncate(before = truncate_num, after = len(group) - truncate_num)
    ##group = group.head(truncate_len)
    ##truncate_len -= truncate_num
    ##group = group.tail(truncate_len)
    ##group = group.fillna(group.median(), inplace = True)
    ##group
    if(count == 0):
        df = group
        count += 1
    else:
        df = pd.concat([df, group])
```

```python
df = df[['property_id', 'prop_type', 'year_built', 'is_new_construction', 'beds', 'baths', 'garage'
        , 'price', 'lot_size', 'building_size', 'line', 'neighborhoods', 'city', 'county', 'state_code'
         , 'baths_full', 'baths_half']]
```

```python
df.head(20)
```

```python
df.insert (3, 'is_multi_family', np.where(df['prop_type'] == 'multi_family', 1, 0))
df.insert (4, 'is_single_family', np.where(df['prop_type'] == 'single_family', 1, 0))
df.insert (5, 'is_condo', np.where(df['prop_type'] == 'condo', 1, 0))
df.drop(columns=['prop_type', 'state_code', 'property_id', 'baths_full', 'baths_half'], inplace = True)
df.loc[(df.is_new_construction == 'f'),'is_new_construction']= 0
df.loc[(df.is_new_construction != False),'is_new_construction']= 1
```

```python
df["baths_beds_ratio"] = df['beds'].where(df['beds'] == 0, round(df['baths']/df['beds'], 4))
df.insert(7, "baths_beds_ratio", df.pop("baths_beds_ratio"))
```
